```python
def get_professionals(data, column):
    data = data.loc[data[column] != 'Student']
    data = data.loc[data[column] != 'Currently not employed']
    data = data.loc[data[column] != 'Not employed']
    data = data.loc[data[column].notna()]
    return data
```
